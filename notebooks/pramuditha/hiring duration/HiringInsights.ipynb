```python
import warnings
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings('ignore')


df = pd.read_csv("cleaned.csv")
```

```python
df['Hired'].value_counts().plot(kind='bar')
```

```python
pd.crosstab(df['Job Title'], df['Hired']).plot(kind='bar', stacked=True)
```

```python
df.groupby('Job Title')['Salary ($1000)'].mean().plot(kind='bar')
```

```python
df.boxplot(column='Salary ($1000)', by='Hired')
```

```python
df['Submission Date'] = pd.to_datetime(df['Submission Date'])
df['Filled'] = pd.to_datetime(df['Filled'])

df['Time_to_Hire'] = (df['Filled'] - df['Submission Date']).dt.days
df['Time_to_Hire'].plot(kind='hist', bins=10)
```

```python
df.boxplot(column='Time_to_Hire', by='Job Title')
```
