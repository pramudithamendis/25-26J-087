```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    text = text.strip()
    return text

# Clean topic and title
df['topic_clean'] = df['topic'].apply(clean_text)
df['title_clean'] = df['title'].apply(clean_text)
```

```python
import pandas as pd

df = pd.read_csv("/content/sample_data/merged_cleaned_dataset.csv")

def extract(col):
    out = []
    for v in df[col].dropna():
        for s in str(v).split(","):
            s = s.strip()
            if s:
                out.append(s)
    return out

skills = sorted(set(extract("skills") + extract("skills_list")))
pd.DataFrame({"skill": skills}).to_csv("unique_skills.csv", index=False)
```
